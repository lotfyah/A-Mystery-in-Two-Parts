```python
# dependencies and setup

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:Password@localhost:5432/EmployeeDB')
connection = engine.connect()
```

```python
# read salary data from SQL
salary_data = pd.read_sql("select * from salaries",connection)
salary_data
```

```python
# histogram to visualize the common salary for employees
plt.hist(salary_data['salary'], density=False, label="salary")
plt.axvline(salary_data['salary'].mean(), color='r', label="average salary")
plt.xlabel("salaries")
plt.ylabel("number of employees")
plt.legend() 
plt.title("salary range")
plt.savefig("salary_range_hist.png")
```

```python
# read title data from SQL
title_data = pd.read_sql("select * from titles", connection)
title_data
```

```python
# read employees data from SQL
employees_data = pd.read_sql("select * from employees", connection)
employees_data
```

```python
# join salary data with employees data then with title data since no direct connection between salary and title
merged_df = pd.read_sql("select * from employees e join salaries s on e.emp_no = s.emp_no join titles t on e.emp_title = t.title_id", connection)
merged_df
```

```python
# average salary by title 
Avg_salary_by_title = merged_df.groupby(['title']).mean()['salary']
Avg_salary_by_title
```

```python
# Bar chart for average salary by title
plot1 = Avg_salary_by_title.plot(kind="bar", title="average salary by title")
plot1.set_xlabel("title")
plot1.set_ylabel("average salary")
plt.savefig("average salary by title.png")
plt.show()
```
